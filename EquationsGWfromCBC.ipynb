# Equations for GW from CBC (inspiral phase)

The aim of this notebook is to compute, plot, and analyze GW from binary systems in the inspiral phase

An inspiral waveform is described by the following nine parameters
- $D$                    distance to the binary
- $m_1$                  mass of the first binary component
- $m_2$                  mass of the second binary component
- $\iota$                inclination angle of the binary
- $\psi$                 polarization angle of the binary
- ($\theta$,$\varphi$)   sky coordinates of the binary
- $t_c$                  end time of the inspiral or the coalescence time
- $\phi_o$               orbital phase of the binary
or?
- $\phi_c$               orbital phase at $t_c$

Where the amplitude ${\cal A}(t)$ is computed as follows:

## ${\cal A}(t) = 4 \left( \frac{G\mathcal{M}}{c^2} \right)^{5/3} \left( \frac{\pi f(t)}{c} \right)^{2/3} $

Note that as the time increases (i.e., the end of the inspiral phase is approached) the frequency and amplitude of the GW increase, therefore the GW signal $h_{+}(t)$ and $h_{\times}(t)$ exhibit the so-called chirp-type waveform.
