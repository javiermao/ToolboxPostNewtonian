# Import libraries

```python
# -------------------------------
# Import packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# -------------------------------
# Define an array for Thetat
Thetat = np.array( [1+2j, 3+4j, 5+6j, 7+8j] )
```

# Compute $\Gamma(1-n,x)=\int_{x}^{\infty}t^{-n}e^{-t}dt = E_n(x)$


<font color='magenta'> Undertanding `mpmath.expint(n,x)`. See description [here](https://mpmath.org/doc/current/functions/expintegrals.html#expint) </font>

```python
# -------------------------------
# Import mpmath.expint
import mpmath as mp
```

```python
# -------------------------------
# Example 1: using 'mpmath.expint' with 'n' and 'x' being a single number (real or complex scalar)
# Note that 'mpmath.expint' can only be evaluated in a single number in 'n' and 'x'
n      = 1/5
x      = -2*1j*np.power(Thetat[0],5/8)/3.5
print("***INPUT***")
print("n          = ", n)
print("x          = ", x)
print(" ")

Fnx     = np.power(x,n-1) * mp.expint(n,x)
FnxReal = float(mp.nstr(Fnx.real, n=8))
FnxImag = float(mp.nstr(Fnx.imag, n=8))
print("***OUTPUT***")
print("Fnx        =   ", Fnx)
print("               ", type(Fnx))
print("real(Fnx)  =   ", FnxReal)
print("               ", type(FnxReal))
print("imag(Fnx)  =   ", FnxImag)
print("               ", type(FnxImag))
print(" ")
```

```python
# -------------------------------
# Example 2: using 'mpmath.expint' with 'n' a single number and 'x' an array
# Note that 'mpmath.expint' cannot only be evaluated in a single number in 'n' and 'x'
n      = 1/5
x      = -2*1j*np.power(Thetat,5/8)/5.3
print("***INPUT***")
print("n        = ", n)
print("           ",type(n))
print("x        = ", x)
print("           ",type(x))
print(" ")

FnxReal = []
FnxImag = []
for x_i in x:
    Fnx_i      = np.power(x_i,n-1) * mp.expint(n,x_i)
    FnxReal.append( float(mp.nstr(Fnx_i.real, n=8)) )
    FnxImag.append( float(mp.nstr(Fnx_i.imag, n=8)) )

FnxReal = np.array( FnxReal )
FnxImag = np.array( FnxImag )
print("***OUTPUT***")
print("FnxReal    =   ", FnxReal)
print("FnxImag    =   ", FnxImag)
print(type(FnxReal))
print(type(FnxImag))
print(" ")

Fnx = FnxReal + 1j*FnxImag
print("Fnx        =   ", Fnx)
print(type(Fnx))
```
